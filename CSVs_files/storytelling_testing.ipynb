```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('movies.csv', lineterminator='\n')
df['profit'] = df['revenue'] - df['budget']

df['genres'] = df['genres'].apply(lambda x: eval(x))

df_separated_genres = df.explode('genres')
df_separated_genres.reset_index(drop=True, inplace=True)
df_separated_genres
```

```python
df_separated_genres_nozero = df_separated_genres.loc[(df_separated_genres['revenue'] != 0) & (df_separated_genres['budget'] != 0)]
japanese_genres_separated = df_separated_genres_nozero[df_separated_genres_nozero['original_language'] == "Japanese"]
japanese_genres_separated
```

```python
no_zeros = df.loc[(df['revenue'] != 0) & (df['budget'] != 0)]

no_zeros
```

```python

japanese_movies = no_zeros[no_zeros['original_language'] == "Japanese"]
japanese_movies
```

```python
jp_descriptive_stats = japanese_movies[['revenue', 'budget']].describe()
# Calculate correlations
jp_correlations = japanese_movies[['revenue', 'budget']].corr()
jp_descriptive_stats, jp_correlations
```

```python
# Calculate correlations
correlations = japanese_movies[['revenue', 'budget']].corr()

# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Scatter plot for selected attributes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='budget', y='revenue', data=df)
plt.title('Scatter Plot of Revenue vs Budget')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df['revenue'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['revenue'], color='skyblue')
plt.title('Box Plot of Revenue')
plt.xlabel('Revenue')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot histograms for revenue and budget
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(japanese_movies['revenue'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(df['budget'], bins=20, color='salmon', edgecolor='black')
plt.title('Histogram of Budget')
plt.xlabel('Budget')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by genre and calculate average metrics
avg_metrics_by_genre = japanese_genres_separated.groupby('genres')[['revenue', 'budget']].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(12, 6))
plt.bar(avg_metrics_by_genre['genres'], avg_metrics_by_genre['revenue'], color='skyblue', label='Revenue')
plt.bar(avg_metrics_by_genre['genres'], avg_metrics_by_genre['budget'], color='salmon', label='Budget')

plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Average')
plt.title('Average Revenue, Budget by Genre')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert release_date to datetime
japanese_movies['release_date'] = pd.to_datetime(japanese_movies['release_date'])

# Extract year from release_date
japanese_movies['year'] = japanese_movies['release_date'].dt.year

# Group by year and calculate average revenue
avg_revenue_by_year = japanese_movies.groupby('year')['revenue'].mean().reset_index()

# Plot line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='revenue', data=avg_revenue_by_year, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Average Revenue')
plt.title('Average Revenue Trend Over the Years')
plt.grid(True)
plt.tight_layout()
plt.show()
```
